```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

COSMOS = pd.read_csv('match_COSMOS_99_with_err_test.csv')
COSMOS_rv = COSMOS.replace(to_replace = -99, value = None)
COSMOS_rv = COSMOS_rv.replace(to_replace = -99.9, value = None)
COSMOS_rv = COSMOS_rv.replace(to_replace = -99.0, value = None)
COSMOS_rv = COSMOS_rv.replace(to_replace = -9.9999, value = None)
print(COSMOS_rv)
```

```python
def HISTO(x,bn=10,df=COSMOS_rv):
    fig = plt.figure(figsize=(10,8))
    plt.hist(df[x],bins='auto')
    plt.xlabel(x,fontsize=25)
    plt.ylabel('counts',fontsize=25)
    plt.xlim(20, 25)
    plt.savefig(x)
    return fig

HISTO('Y_MAG_APER2')
# HISTO('ip_MAGERR_APER2')
HISTO('ip_MAG_APER2')
# HISTO('r_MAGERR_APER2')
HISTO('r_MAG_APER2')
# HISTO('yHSC_MAGERR_APER2')
HISTO('yHSC_MAG_APER2')
# HISTO('zpp_MAGERR_APER2')
HISTO('zpp_MAG_APER2')
# HISTO('zspec')
```
